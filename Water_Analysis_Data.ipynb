```python
df["ph"].fillna(df["ph"].mean(),inplace=True)
```

```python
df["Sulfate"].fillna(df["Sulfate"].median(),inplace=True)
```

```python
df["Trihalomethanes"].fillna(df["Trihalomethanes"].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
# Distribution of numerical features in train data
df.hist(figsize=(12, 10), bins=30)
plt.suptitle("Distribution of Numerical Features in Water Potability")
plt.show()
```

```python
numeric_df=df.select_dtypes(include=["float","int"])
```

```python
numeric_df.boxplot(figsize=(12, 6))
plt.xticks(rotation=45)
plt.title("Boxplot of Numerical Features in water potability")
plt.show()
```

```python
corr_max=numeric_df.corr()
```

```python
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.heatmap(corr_max, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap - Water Potability")
plt.show()
```
